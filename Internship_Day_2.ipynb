```python
login_logs = {
    "user1": 3,
    "user2": 17,
    "user3": 5,
    "user4": 23,
    "user5": 11,
    "user6": 2,
    "user7": 19
}

count = 0
user_names = []
attempts_list = []
risk_levels = []
suspicious_user = []

for user, attempts in login_logs.items():
    user_names.append(user)
    attempts_list.append(attempts)

    if attempts <= 5:
        risk = "Low Risk"
    elif attempts <= 10:
        risk = "Medium Risk"
    else:
        risk = "High Risk"
        count += 1
        suspicious_user.append(user)
    risk_levels.append(risk)

suspicious_user_percentage = (count / len(login_logs)) * 100
print("Suspicious users : ",suspicious_user)
print("The percentage of suspicious users:", suspicious_user_percentage, "%")
print("\nSecurity Report:")

for i in range(len(user_names)):
    print("UserName:", user_names[i],"| Attempts:", attempts_list[i],"| Risk Level:", risk_levels[i])
```

```python
import pandas as pd

data = {
    "User": ["Amit", "Riya", "Sohan", "Priya", "Ankit", "Rahul", "Sneha"],
    "Failed Attempts": [3, 15, 4, 21, 8, 12, 2],
    "Location": ["India", "Russia", "India", "Unknown", "India", "China", "India"]
}

df = pd.DataFrame(data)

print("All Records:")
print(df)

suspicious_user = []
suspicious_locations_user = []
threat_level = []

for index, row in df.iterrows():
    if row["Failed Attempts"] <= 5:
        risk = "Low Risk"
    elif row["Failed Attempts"] <= 10:
        risk = "Medium Risk"
    else:
        risk = "High Risk"
        suspicious_user.append(row["User"])

    threat_level.append(risk)

    # Suspicious Location
    if row["Location"] == "Unknown":
        suspicious_locations_user.append(row["User"])

df["Threat_Level"] = threat_level

print("\nSuspicious Users:")
print(suspicious_user)

print("\nUsers from Suspicious Locations:")
print(suspicious_locations_user)

print("\nDataFrame with Threat Level:")
print(df)

# Create suspicious records DataFrame
suspicious_records = []

for index, row in df.iterrows():
    if row["Threat_Level"] == "High Risk" or row["Location"] == "Unknown":
        suspicious_records.append(row)

suspicious_df = pd.DataFrame(suspicious_records)

print("\nSuspicious Records:")
print(suspicious_df)

suspicious_df.to_csv("suspicious_records.csv", index=False)

print("\nSuspicious records exported to 'suspicious_records.csv'")
```

```python
import pandas as pd

network_logs = [
{"IP":"192.168.1.1", "Packets": 120},
{"IP": "192.168.1.2", "Packets": 850},
{"IP": "192.168.1.3", "Packets": 95} ,
{"IP": "192.168.1.4", "Packets": 1200},
{"IP":"192.168.1.5", "Packets": 200}
]

df = pd.DataFrame(network_logs)
print(df)
threshold = 500
suspicious_ip = []
count=0
sum_packet=0
for index,row in df.iterrows():
    if row["Packets"] > threshold:
        suspicious_ip.append(row["IP"])
        count=count+1
        sum_packet=sum_packet+row["Packets"]
    else:
        sum_packet=sum_packet+row["Packets"]

print("\nSuspicious IP addresses unusually high traffic :")
print(suspicious_ip)

for index,row in df.iterrows():
    if row["Packets"] > threshold:
      print("Alert! Suspicious ip :",row["IP"],"with packets :",row["Packets"])

avg_packet=sum_packet/len(df)

print("The number of suspicious IPs : ",count)
print("The average packet traffic  : ",avg_packet)


# Create suspicious records DataFrame
suspicious_ips = []

for index, row in df.iterrows():
    if row["Packets"] > threshold :
        suspicious_ips.append(row)

suspicious_df = pd.DataFrame(suspicious_ips)

print("\nSuspicious Records:")
print(suspicious_df)

suspicious_df.to_csv("suspicious_records.csv", index=False)

print("\nSuspicious records exported to 'suspicious_records.csv'")
print("\nNetwork Threat Report : ")
print("Total Devices Monitored : ",len(df))
print("Total Suspicious Devices : ",count)
print("Average Packet Traffic : ",avg_packet)

print("\nRecommended Security Actions:")
print("1. Investigate suspicious IP addresses generating high traffic.")
print("2. Review network and system logs regularly.")
print("3. Keep systems updated with the latest security patches.")
print("4. Use strong authentication and access controls.")
print("5. Perform regular security audits and vulnerability scans.")
print("6. Configure automatic alerts for abnormal traffic.")
print("7. Maintain secure backups of important data.")
```
