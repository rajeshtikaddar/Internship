**Phishing Email Detection System**

Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix as sk_confusion_matrix
)
```

**Load DataSet**

```python
# Create the dataset
data = {
    "Subject_Length": [45, 18, 52, 20, 48],
    "Contains_Link": [1, 0, 1, 0, 1],
    "Contains_Urgent_Words": [1, 0, 1, 0, 0],
    "Sender_Reputation": [2, 9, 3, 8, 4],
    "Phishing": ["Yes", "No", "Yes", "No", "Yes"]
}

# Load dataset into a DataFrame
df = pd.DataFrame(data)

# Display the dataset
print(df)
```

**Explore the DataSet**

```python
# First 5 rows
print(df.head())

# Dataset information
print(df.info())

# Statistical summary
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Check class distribution

print(df["Phishing"].value_counts())
```

**Select Features and Target Variable**

```python
# Features (X) and Target Variable (y)
X = df[
[
    "Subject_Length",
    "Contains_Link",
    "Contains_Urgent_Words",
    "Sender_Reputation"]]

y = df["Phishing"]
```

**Split Data into Training and Testing Sets**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

**Logistic regreassion**

```python
from sklearn.linear_model import LogisticRegression

# Create and train the model
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

**Decision Tree**

```python
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier

# Create and train the model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

**Predictions **

```python
y_pred = dt_model.predict(X_test)
```

**Evaluate the Model**

```python
# Accuracy
accuracy= accuracy_score(y_test, y_pred)
# Presicion
precision =precision_score(y_test, y_pred,pos_label='Yes', zero_division=0)
#Recall
recall = recall_score(y_test, y_pred,pos_label='Yes', zero_division=0)
#F1 Score
f1 = f1_score(y_test, y_pred,pos_label='Yes', zero_division=0)

#confusion matrix
conf_matrix = sk_confusion_matrix(y_test, y_pred, labels=['No', 'Yes'])

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"Confusion Matrix: {conf_matrix}")
```

**Prediction Challenge**

```python
new_email = pd.DataFrame({
  'Subject_Length': [55],
  'Contains_Link': [1],
  'Contains_Urgent_Words': [1],
  'Sender_Reputation': [2]
})
prediction = dt_model.predict(new_email)

print("\n === PREDICTION CHALLENGE === ")
print("New Email Prediction:", prediction[0])
if prediction[0] == "Yes":
  print("Result: Phishing Email")
else:
  print("Result: Not Phishing")
```
