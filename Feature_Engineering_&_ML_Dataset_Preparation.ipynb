Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "Failed_Logins": [2, 18, 1, 22, 3, 15],
    "Login_Time": ["10:00 AM", "2:00 AM", "11:00 AM", "3:15 AM", "9:30 AM", "2:45 AM"],
    "Country": ["India", "Russia", "India", "Unknown", "India", "China"],
    "Device_Type": ["Laptop", "Desktop", "Mobile", "Laptop", "Desktop", "Mobile"],
    "Suspicious_Activity": ["No", "Yes", "No", "Yes", "No", "Yes"]
}

df = pd.DataFrame(data)
print(df)
```

**DataSet Exploration**

```python
df.info()
df.describe()
```

**Feature Engineering **

```python
high_risk = ["Russia","China","Unknown"]
df["High_Risk_Country"] = df["Country"].apply(
  lambda x: 1 if x in high_risk else 0
)
```

```python
df.head()
```

Login Time

```python
df["Hour"] = pd.to_datetime(
    df["Login_Time"],
    errors="coerce"
).dt.hour

df["Night_Login"] = (df["Hour"] < 6 ).astype(int)

df[["Login_Time","Hour"]].head(10)
```

```python
df.head()
```

**Create Login Risk Score **


```python
df["Login_Risk_Score"] = (
    df["Failed_Logins"]*0.6 +
    df["High_Risk_Country"]*20 +
    df["Night_Login"]*10
)
```

```python
df.head()
```

**Encoding Categorial Features **

```python
from sklearn.preprocessing import LabelEncoder

le =  LabelEncoder()

df["Device_Type"] = le.fit_transform(
    df["Device_Type"]
)

df["Suspicious_Activity"] = le.fit_transform(
    df["Suspicious_Activity"]
)

df.head()
```

**Feature Selections And Final Dataset **

```python
final_dataset = df[
    [
    "Failed_Logins",
    "Device_Type",
    "High_Risk_Country",
    "Night_Login",
    "Login_Risk_Score",
    "Suspicious_Activity"
    ]
]

final_dataset.head()
```

Target Variable: Suspicious_Activity

Input Features: Failed_Logins, Country, Device_Type, High_Risk_Country, Night_Login, Login_Risk_Score, Hour

Most Important Features Failed_Logins, High_Risk_Country, Night_Login, Login_Risk_Score
